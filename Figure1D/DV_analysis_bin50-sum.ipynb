```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_file_name = pd.read_csv("DV_distribution/file_name.csv", header=None)
data_file_name.columns=['short', 'long']
```

```python
right_file_name = []
left_file_name = []
right_id = []
left_id = []

_3ss_file_name = []
_4ss_file_name = []
_5ss_file_name = []
_6ss_file_name = []
_7ss_file_name = []
_8ss_file_name = []

_3ss_file_id = []
_4ss_file_id = []
_5ss_file_id = []
_6ss_file_id = []
_7ss_file_id = []
_8ss_file_id = []

for id_, data in data_file_name.iterrows():
    if data.long[-5] == 'l':
        left_file_name.append(data.long)
        left_id.append(data.short)
    if data.long[-5] == 'r':
        right_file_name.append(data.long)
        right_id.append(data.short)
        
    if data.long[0] == '3':
        _3ss_file_name.append(data.long)
        _3ss_file_id.append(data.short)
    if data.long[0] == '4':
        _4ss_file_name.append(data.long)
        _4ss_file_id.append(data.short)
    if data.long[0] == '5':
        _5ss_file_name.append(data.long)
        _5ss_file_id.append(data.short)
    if data.long[0] == '6':
        _6ss_file_name.append(data.long)
        _6ss_file_id.append(data.short)
    if data.long[0] == '7':
        _7ss_file_name.append(data.long)
        _7ss_file_id.append(data.short)
    if data.long[0] == '8':
        _8ss_file_name.append(data.long)
        _8ss_file_id.append(data.short)
        
```

```python
tmp_id = [] 
for id_, data in data_file_name.iterrows():
    if (data.long.split('_')[0] == '7ss') & (data.long.split('_')[1]=='0'):
        tmp_id.append(data.short)  
data_all_embryo[list(set(left_id).intersection(tmp_id))].sum(axis=1).to_numpy()
```

```python
somite_stage = ["3ss", "4ss", "5ss", "6ss", "7ss", "8ss"]
n_embryo = list('01234')
n_embryo.append('2.1')

data_right = {}
data_left = {}

#Right
for s in somite_stage:
    for n in n_embryo:
        tmp_id = [] 
        for id_, data in data_file_name.iterrows():
            if (data.long.split('_')[0] == s) & (data.long.split('_')[1] == n):
                tmp_id.append(data.short)  
        if len(tmp_id)>0:
            data_right[s+'_'+n]=data_all_embryo[list(set(right_id).intersection(tmp_id))].sum(axis=1).to_numpy()
            
data_all_embryo_sum_right = pd.DataFrame(data=data_right)


#Left
for s in somite_stage:
    for n in n_embryo:
        tmp_id = [] 
        for id_, data in data_file_name.iterrows():
            if (data.long.split('_')[0] == s) & (data.long.split('_')[1] == n):
                tmp_id.append(data.short) 
        if len(tmp_id)>0:
            data_left[s+'_'+n]=data_all_embryo[list(set(left_id).intersection(tmp_id))].sum(axis=1).to_numpy()
            
data_all_embryo_sum_left = pd.DataFrame(data=data_left)
```
